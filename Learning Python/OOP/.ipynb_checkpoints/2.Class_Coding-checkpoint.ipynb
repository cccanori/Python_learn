___Class Coding Basics___

__Classes Generate Multiple Instance Objects__

```python
'''- Class objects: provide default behavior and serve as factories for instance objects.
     They come from statments
   - Instance objects: are the real objects your programs process—each is a namespace in
     its own right, but inherits (i.e., has automatic access to) names in the class from which
     it was created.They come from calls'''
#By contrast, modules are only one objectd copied from a file --> cannot support multiple objects(instances)
# With classes, each instance can have its own featurres and support the classs behavior.
'''Thats why OOP and functional programing are so related (Factories-Clouserues allow customization by using the scope of a function).
   Nevertheless, they are not mutually exclusive'''
```

_Class Objects Provide Default Behavior (statment)_

```python
'''-The class statement creates a class object and assigns it a name: Generates a calass object assiciated with the header variable
    in the class statent
    
   -Assignments inside class statements make class attributes: Each top level assignment (not nested) within the class code,
    morph into class object attributes of the form object.module
    
   -Class attributes provide object state and behavior: All class-object attributes recorded generate methods for 
    future instances'''
```

_Instance Objects Are Concrete Items (call)_

```python
''' • Calling a class object like a function makes a new instance object.Each time a class is called, it creates and returns a new instance object
    
    • Each instance object inherits class attributes and gets its own namespace. Instance objects created from classes are new namespaces; 
      they start out empty but inherit attributes that live in the class objects from which they were generated.
    
    • Assignments to attributes of self in methods make per-instance attributes: Inside a class’s method functions, 
      the first argument (called self by convention) references the instance object being processed; assignments to attributes of
      self create or change data in the instance, not the class.'''
```

```python
'''Class and instance coding flow'''

#1. Setting a class:
class FirstClass: # Define a class object --> name of the class
    def setdata(self, value): # Define class's methods --> funct defines the behavior of the class
        self.data = value # self is the instance
    def display(self):
        print(self.data) # self.data: per instance
'''remember that class attribute (top-level script in the class) != methods (def in the class)'''
# functions inside classes (methods) support everything seen so far: defaults, yields, so on...

#2. Setting instances
>>> x = FirstClass() # Make two instances
>>> y = FirstClass() # Each is a new namespace
''' x and y (instances) are namespaces that have acces to FirstClass attributes. At first they are empty spaces (x and y)'''

#3. calling the methods:
>>> x.setdata("King Arthur") # Call methods: self is x
>>> y.setdata(3.14159) # Runs: FirstClass.setdata(y, 3.14159)
'''x and y are no more empty spaces of data --> notice that the data is store in the instance (thx to delf argument )
   and not in the class itself
   Notice that different types of data structures (string and float) were stored in each instance, but not in the class'''

#4. Same method with differnt result (a class can hold multiple instances)
>>> x.display() # self.data differs in each instance
King Arthur
>>> y.display() # Runs: FirstClass.display(y)
3.14159

#5. assigning the attribute outside the method: After the setdata() attribute was executed!
>>> x.data = "New value" # Can get/set attributes
>>> x.display() # Outside the class too
New value
'''This is call explicit assigning object'''
# we can even generate a completely new instance-object by outside assignment:
>>> x.anothername = "spam" # Can set new attributes here too!
'''anothername is only linked to the instance namesapce and not the class-object namespace--> privacy code based '''

```

__Classes Are Customized by Inheritance__

```python
'''Besides serving as factories for generating multiple instance objects, classes also allow us to make changes by 
   introducing new components (called subclasses), instead of changing existing components in place.'''
# This is called hierarchies of classes !!! --> classes (sub) that inherits attributes from other classes (super)

'''* Hierarchy main use: OVERRIDE the more general defitions in the class-tree--> the more further down the class is in the searching-tree
   the more specific thesoftware becomes '''

'''• Superclasses are listed in parentheses in a class header.
   • Classes inherit attributes from their superclasses.
   • Instances inherit attributes from all accessible classes: Each instance gets
     names from the class it’s generated from, as well as all of that class’s superclasses.
   • Each object.attribute reference invokes a new, independent search 
   • Logic changes are made by subclassing, not by changing superclasses 
   
This logic (class inheritance) make OOP to support factoring and customization of code better than any other paradigm
we have seen so far --> There is still to see new-style descriptors and metaclasses !! --> make it even more powerful'''
```

```python
# Rol of inheritance:
class FirstClass: # Define a class object --> name of the class
    def setdata(self, value): # Define class's methods --> funct defines the behavior of the class
        self.data = value # self is the instance
    def display(self):
        print(self.data) # self.data: per instance
        
class SecondClass(FirstClass): # Inherits setdata
    def display(self): # Changes display --> OVERRIDES the display class-attribute superclass --> sth called overloading
        print('Current value = "%s"' % self.data) # but its completely independent to first class

# Setting an instance:
>>> z = SecondClass()
>>> z.setdata(42) # Finds setdata in FirstClass
>>> z.display() # Finds overridden method in SecondClass
Current value = "42" # prints the second class' attribute

>>> x.display() # x is still a FirstClass instance (old message) --> the second class doesn't affect the first class object set before
New value       # we customized it
```

__Classes Can Intercept Python Operators: Operators Overloading__

```python
'''Operator overloading: It allows object-classes can intercept and resopnd to built-in operators such as addition, 
   slicing, printing,etc. Main charachteristics:
   * Methods named with double underscores (__X__) are special hooks: In Python
     classes we implement operator overloading by providing specially named methods
     to intercept operations --> fixed and unchangeable mapping from each operator (built-in) to a named method (__X__).
   
   * Such methods are called automatically when instances appear in built-in operations.For instance, when an instance faces a 
      + operand, __add__ will be called (wether was inherited or assigned to the instance)
   
   * Classes may override most built-in type operations. There are dozens of special operator overloading method 
     names for intercepting and implementing nearly every operation available for built-in types. It means, there are almost
     enough operators overloaders that supply the wholy built-in functions.
   
   * There are no defaults for operator overloading methods, and none are required. A class facing an operand (e.g, +) and
     the operator __add__ wasnt set before, python will raise an exception.
     
   * New-Style (python 3.X) classes have some defaults. some classes named object bring up some __X__ methods. However, 
     they are not many and common.
    
    * Operators allow classes to integrate with Python’s object model. By overloading type operations, the user-defined objects 
      we implement with classes can act just like built-ins, and so provide consistency as well as compatibility with expected interfaces.
      Which is, user set objects can act as built-ins in other inteerfaces'''

# Operator overloading is optional: Primarly used for programming developing tools for other python programmers.
# but there is one which is really common: __init__ --> the constructor method --> as important as self argument in OOP design

```

```python
'Using previous examples': 
    
class FirstClass: # Define a class object --> name of the class
    def setdata(self, value): # Define class's methods --> funct defines the behavior of the class
        self.data = value # self is the instance
    def display(self):
        print(self.data) # self.data: per instance
        
class SecondClass(FirstClass): # Inherits setdata
    def display(self): # Changes display --> OVERRIDES the display class-attribute superclass --> sth called overloading
        print('Current value = "%s"' % self.data) # but its completely independent to first class

class ThirdClass(SecondClass): # Inherit from SecondClass. Hence, inherits from FirstClass as well.
    def __init__(self, value): # On "ThirdClass(value)"
        self.data = value #Notice that __init__ doesnt has a return statment: 
    def __add__(self, other): # On "self + other" --> allowas '+' addition
        return ThirdClass(self.data + other)
    def __str__(self): # On "print(self)", "str()" --> allows a printing or str statment method
        return '[ThirdClass: %s]' % self.data
    def mul(self, other): # In-place change: named
        self.data *= other

#Now let's look the program in action:

>>> a = ThirdClass('abc') # __init__ called
>>> a.display() # Inherited method called, this is from SecondClass
Current value = "abc"
>>> print(a) # __str__: returns display string
[ThirdClass: abc]

>>> b = a + 'xyz' # __add__: makes a new instance
>>> b.display() # b has all ThirdClass methods
Current value = "abcxyz"
>>> print(b) # __str__: returns display string
[ThirdClass: abcxyz]

>>> a.mul(3) # mul: changes instance in place --> changes self.data itself
>>> print(a) #Notice that a is being changed instead of b
[ThirdClass: abcabcabc]
```

```python
'''Why Operator Overloading: how much you want your object to look and feel like built-in types.
    
    Frankly, many operator overloading methods tend to be used only when you are implementing
    objects that are mathematical in nature; a vector or matrix class may overload
    the addition operator, for example, but an employee class likely would not. For
    simpler classes, you might not use overloading at all, and would rely instead on explicit
    method calls to implement your objects’ behavior.'''
```

__Records Revisited: Classes Versus Dictionaries__

```python
# first type of record
>>> rec = ('Bob', 40.5, ['dev', 'mgr']) # Tuple-based record
>>> print(rec[0])
Bob

#Second class of record:
>>> rec = {}
>>> rec['name'] = 'Bob' # Dictionary-based record
>>> rec['age'] = 40.5 # Or {...}, dict(n=v), etc.
>>> rec['jobs'] = ['dev', 'mgr']
>>>
>>> print(rec['name'])
Bob

# Doing the same with classes:
>>> class rec: pass
>>> rec.name = 'Bob' # Class-based record
>>> rec.age = 40.5
>>> rec.jobs = ['dev', 'mgr']
>>> print(rec.name)
Bob
'''However, this form affects the class. So its more paractical to make instances, and then add the records :'''
>>> class rec: pass
>>> pers1 = rec() # Instance-based records
>>> pers1.name = 'Bob' # Noe the records are at the instance level only
>>> pers1.jobs = ['dev', 'mgr']
>>> pers1.age = 40.5
>>>
>>> pers2 = rec()
>>> pers2.name = 'Sue'
>>> pers2.jobs = ['dev', 'cto']
>>>
>>> pers1.name, pers2.name # each instannce has a unique attribute name
('Bob', 'Sue')

# Now let's implement a ful-blown OOP class that proccedural programming records cannot support (directly):
class Person:
    def __init__(self, name, jobs, age=None): # class = data + logic --> colect data as the class is called!!
        self.name = name
        self.jobs = jobs
        self.age = age
    def info(self):
        return (self.name, self.jobs) # calsses allow us to store functions in the object-namespace --> isnt natural in dicts and uples
                                      # Even if they can actually be stored
>>> rec1 = Person('Bob', ['dev', 'mgr'], 40.5) # Construction calls
>>> rec2 = Person('Sue', ['dev', 'cto'])
>>>
>>> rec1.jobs, rec2.info() # Attributes + methods
(['dev', 'mgr'], ('Sue', ['dev', 'cto']))
```

```python
'''Important things to bare in mind dealing with classes and instances:'''

>>> class rec: pass # creating a class by statment --> iT doesnt have any attribute at this statment

>>> rec.name = 'Bob' # Just objects with attributes --> Obj attributes (class methods) could be added outside the class statment!!! 
>>> rec.age = 40

>>> x = rec() # Instances inherit class names --> both x and y inherited .name and .age attributes from rec class
>>> y = rec()
>>> x.name, y.name # name is stored on the class only
('Bob', 'Bob')

>>> x.name = 'Sue' # But assignment changes x only --> an instance change in the attribut doesnt affect the class methods --> Overrides the inherited data
>>> rec.name, x.name, y.name # same method for different object-namespaces
('Bob', 'Sue', 'Bob')
%---------------------------------------------------------------------
>>> x.name, x.__dict__['name'] # Attributes present here are dict keys
('Sue', 'Sue')
>>> x.age # But attribute fetch checks classes too
40
>>> x.__dict__['age'] # Indexing dict does not do inheritance --> dictionaries dont support inheritance while classes do!!!
KeyError: 'age'

% ----------------------------------------------------------------------

>>> x.__class__ # Instance to class link --> .__class__ this method shows the main class behind the object
<class '__main__.rec'>

>>> rec.__bases__ # Class to superclasses link, () in 2.X --> .__bases__ drops a tuple with the superclasses of a class
(<class 'object'>,)

% -------------------------------------------------------------------------------
#creating a single function apart:
>>> def uppername(obj):
    return obj.name.upper() # Still needs a self argument (obj)
>>> uppername(x) # Call as a simple function
'SUE'

>>> rec.method = uppername # Now it's a class's method! --> assigning a function name (not the call!) to a class-method
                           # makes a new method with the name of the function !!!!
>>> x.method() # Run method to process x
'SUE'
>>> y.method() # Same, but pass y to self
'BOB'
>>> rec.method(x) # Can call through instance or class
'SUE'
% -------------------------------------------------------------------

remember that self respresents the automatic object argument --> cornerstone of python OOP !!!!!!!
While __init__ is really widespread across OOP --> its mandatory to get its meaning

Name spaces (searching-tree) are important to support self argument, but isnt the corner stone. 
```

```python
# Types of class-objects:
'''Super:
   Defines a method function and a delegate that expects an action in a subclass.
Inheritor:
   Doesn’t provide any new names, so it gets everything defined in Super.
Replacer:
   Overrides Super’s method with a version of its own.
Extender:
    Customizes Super’s method by overriding and calling back to run the default.
Provider:
    Implements the action method expected by Super’s delegate method.'''

class Super:
    def method(self):
        print('in Super.method') # Default behavior
    def delegate(self):
        self.action() # Expected to be defined
class Inheritor(Super): # Inherit method verbatim
    pass
class Replacer(Super): # Replace method completely
    def method(self):
        print('in Replacer.method')
class Extender(Super): # Extend method behavior
    def method(self):
        print('starting Extender.method')
        Super.method(self)
        print('ending Extender.method')

class Provider(Super): # Fill in a required method --> In this case, Super is a ABSTRACT superclass:
    def action(self):   # A super class' behavior that is provided by a subclass 
        print('in Provider.action')

if __name__ == '__main__':
for klass in (Inheritor, Replacer, Extender):
print('\n' + klass.__name__ + '...')
klass().method()
print('\nProvider...')
x = Provider()
x.delegate()

# After ran as main script:
Inheritor...
in Super.method
Replacer...
in Replacer.method
Extender...
starting Extender.method
in Super.method
ending Extender.method
Provider...
in Provider.action
```

```python
'''Name scope example: All x are completelly different in each declaration'''
# manynames.py
X = 11 # Global (module) name/attribute (X, or manynames.X)
def f():
    print(X) # Access global X (11)
def g():
    X = 22 # Local (function) variable (X, hides module X)
    print(X)
class C:
    X = 33 # Class attribute (C.X)
    def m(self):
        X = 44 # Local variable in method (X)
        self.X = 55 # Instance attribute (instance.X)
if __name__ == '__main__':
    print(X) # 11: module (a.k.a. manynames.X outside file)
    f() # 11: global
    g() # 22: local
    print(X) # 11: module name unchanged
    obj = C() # Make instance
    print(obj.X) # 33: class name inherited by instance
    obj.m() # Attach attribute name X to instance now
    print(obj.X) # 55: instance
    print(C.X) # 33: class (a.k.a. obj.X if no X in instance)
    #print(C.m.X) # FAILS: only visible in method
    #print(g.X) # FAILS: only visible in function
        
'''When the scopes are imported:'''

# otherfile.py
import manynames
X = 66
print(X) # 66: the global here
print(manynames.X) # 11: globals become attributes after imports
manynames.f() # 11: manynames's X, not the one here!
manynames.g() # 22: local in other file's function
print(manynames.C.X) # 33: attribute of class in other module
I = manynames.C()
print(I.X) # 33: still from class here
I.m()
print(I.X) # 55: now from instance!

'''A brief summary of fun() scope:using global and nonlocal expressions'''
X = 11 # Global in module
def g1():
    print(X) # Reference global in module (11)
def g2():
    global X
    X = 22 # Change global in module
def h1():
    X = 33 # Local in function
def nested():
    print(X) # Reference local in enclosing scope (33)
    def h2():
        X = 33 # Local in function
def nested():
    nonlocal X # Python 3.X statement
    X = 44 # Change local in enclosing scope

    
    
'''Nested scopes: when classes are inside closures'''
X = 1
def nester():
    X = 2 # Hides global
    print(X) # Local: 2
    class C:
        X = 3 # Class local hides nester's: C.X or I.X (not scoped) --> LEGB rulo not CLEGB rule (C is class) !
        print(X) # Local: 3
        def method1(self):
            print(X) # In enclosing def (not 3 in class!): 2
            print(self.X) # Inherited class local: 3
        def method2(self):
            X = 4 # Hides enclosing (nester, not class)
            print(X) # Local: 4
            self.X = 5 # Hides class
            print(self.X) # Located in instance: 5
    I = C()
    I.method1()
    I.method2()

print(X) # Global: 1
nester() # Rest: 2, 3, 2, 3, 4, 5
print('-'*40)
```

__Name spaces: dictionaries__ 

```python
'''__class__ and __bases__ allow you to inspect inheritance hierarchies in your objects '''

>>> class Super:
        def hello(self):
            self.data1 = 'spam'
>>> class Sub(Super):
        def hola(self):
            self.data2 = 'eggs'
            
>>> X = Sub()
>>> X.__dict__ # Instance namespace dict
{}
>>> X.__class__ # Class of instance --> what class the obj was created with !
<class '__main__.Sub'>
>>> Sub.__bases__ # Superclasses of class --> returns a tuple with the superclasses in the set order
(<class '__main__.Super'>,)
>>> Super.__bases__ # () empty tuple in Python 2.X --> cause Super doesnt have bases
(<class 'object'>,)

>>> Y = Sub()
>>> X.hello()
>>> X.__dict__
{'data1': 'spam'}
>>> X.hola()
>>> X.__dict__
{'data2': 'eggs', 'data1': 'spam'}

>>> list(Sub.__dict__.keys())
['__qualname__', '__module__', '__doc__', 'hola'] # you see the class' methods not thedata --> thats oly in the instance
>>> list(Super.__dict__.keys())
['__module__', 'hello', '__dict__', '__qualname__', '__doc__', '__weakref__']
>>> Y.__dict__
{}

'''Because attributes are actually dictionary keys inside Python, there are really two ways
to fetch and assign their values—by qualification, or by key indexing:'''

>>> X.data1, X.__dict__['data1']
('spam', 'spam')
>>> X.data3 = 'toast' # Qualification
>>> X.__dict__
{'data2': 'eggs', 'data3': 'toast', 'data1': 'spam'}
>>> X.__dict__['data3'] = 'ham' #Indexing
>>> X.data3
'ham'
```
