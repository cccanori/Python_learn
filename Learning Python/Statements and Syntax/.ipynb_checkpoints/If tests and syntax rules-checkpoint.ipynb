___Truth tests___:
- Any nonzero number or nonempty object is true.
- Zero numbers, empty objects, and the special object None are considered false.
- Comparisons and equality tests are applied recursively to data structures.
- Comparisons and equality tests return True or False (custom versions of 1 and 0).
- Boolean and and or operators return a true or false operand object.

```python
X and Y # true if both are true
X or Y # true if one is true
not X # true if X is false
```

```python
# OR case:
>>> 2 or 3, 3 or 2 # Return left operand if true
(2, 3) # Else, return right operand (true or false)
>>> [] or 3
3
>>> [] or {}
{}
# AND case:
>>> 2 and 3, 3 and 2 # Return rigth operand if false
(3, 2) # Else, return right operand (true or false)
>>> [] and {}
[]
>>> 3 and []
[]
```

```python
# SHOR-CIRCUIT IF:
if X:
A = Y
else:
A = Z # Large circuit

A = Y if X else Z # SHORT CIRCUIT --> The same as above!!
#e.g.,:
>>> A = 't' if 'spam' else 'f' # Nonempty is true
>>> A
't'
>>> A = 't' if '' else 'f'
>>> A
'f'
A = ((X and Y) or Z) # same short circuit using and & or

A = [Z, Y][bool(X)] # same result using lists
#e.g,.
>>> ['f', 't'][bool('')] # bool of '' is false, then 0
'f'
>>> ['f', 't'][bool('spam')] # bool of 'spam' is True, then 1
't'
```

Break,continue,pass, and else:

```python
break
Jumps out of the closest enclosing loop (past the entire loop statement)
continue
Jumps to the top of the closest enclosing loop (to the loop’s header line)
pass
Does nothing at all: it’s an empty statement placeholder
Loop else block 
Runs if and only if the loop is exited normally (i.e., without hitting a break)

# general format:
while <test1>:
    <statements1>
    if <test2>: break # Exit loop now, skip else
    if <test3>: continue # Go to top of loop now, to test1
else:
    <statements2> # Run if we didn't hit a 'break'
```

__FOR:__ Step across any kind of sequence object 

```python
# generic form
for <target> in <object>: # Assign object items to target
    <statements> # Repeated loop body: use target
else:
    <statements> # If we didn't hit a 'break' --> executes if all items have been iterated

# mixing it with break and continue:
for <target> in <object>: # Assign object items to target
    <statements>
    if <test>: break # Exit loop now, skip else
    if <test>: continue # Go to top of loop now
else:
    <statements> # If we didn't hit a 'break'
    
    
# Ex:
sum = 0
for x in [1, 2, 3, 4]:
    sum = sum + x
sum
10

prod = 1
for item in [1, 2, 3, 4]: prod *= item
prod
24

# Other types of sequences (sequences are iterables but not all iterables are sequences: e.g,dicts and files)
>>> S = "lumberjack"
>>> T = ("and", "I'm", "okay")
     for x in S: print(x, end=' ') # Iterate over a string
l u m b e r j a c k
>>> for x in T: print(x, end=' ') # Iterate over a tuple
and I'm okay
```

```python
# tuple assignment for loops: Tuple unpacking case !!
>>> T = [(1, 2), (3, 4), (5, 6)]
>>> for (a, b) in T: # Tuple assignment at work --> the loop target can be a tuple if you are iterating tuples
        print(a, b)
1 2 # first step: (a,b) = (1,2)
3 4 # and so on...
5 6
# Normally used with zip call to implement paralell traversals
# Often used with SQL in python: the rows come in tuples in a list (DB)
```

```python
# tuple assignment for dicts: Tuple unpacking case in keys and values !!
#for using keys and indexes
>>> D = {'a': 1, 'b': 2, 'c': 3}
>>> for key in D:
... print(key, '=>', D[key]) # Use dict keys iterator and index
...
a => 1
c => 3
b => 2
# Same for using tuple unpacking:
>>> list(D.items())
[('a', 1), ('c', 3), ('b', 2)]
>>> for (key, value) in D.items():
... print(key, '=>', value) # Iterate over both keys and values
...
a => 1
c => 3
b => 2

# Same version using manual unpacking: on extra step
>>> T
[(1, 2), (3, 4), (5, 6)]
>>> for both in T:
... a, b = both # Manual assignment equivalent
... print(a, b)
...
1 2
3 4
5 6

# Nested sequences work too:
#E.g.,
>>> ((a, b), c) = ((1, 2), 3) # Nested sequences work too
>>> a, b, c
(1, 2, 3)
>>> for ((a, b), c) in [((1, 2), 3), ((4, 5), 6)]: print(a, b, c)
...
1 2 3
4 5 6

>>> for ((a, b), c) in [([1, 2], 3), ['XY', 6]]: print(a, b, c)
...
1 2 3
X Y 6

#Using extended (starred*) assignment:
>>> a, *b, c = (1, 2, 3, 4) # Extended seq assignment
>>> a, b, c
(1, [2, 3], 4)
>>> for (a, *b, c) in [(1, 2, 3, 4), (5, 6, 7, 8)]:
... print(a, b, c)
...
1 [2, 3] 4 # * Always returns a list !!
5 [6, 7] 8

# same but using slicing:
>>> for all in [(1, 2, 3, 4), (5, 6, 7, 8)]: # Manual slicing in 2.6
... a, b, c = all[0], all[1:3], all[3]
... print(a, b, c)
1 (2, 3) 4
5 (6, 7) 8 # In this case, there aren't a list!!!
```

```python
# Nested for loops:

>>> items = ["aaa", 111, (4, 5), 2.01] # A set of objects
>>> tests = [(4, 5), 3.14] # Keys to search for
>>>
for key in tests: # For all keys
    for item in items: # For all items
             if item == key: # Check for match
                print(key, "was found")
                break
    else:                       # this else is linked to the inner for, not the outer 
        print(key, "not found!")

(4, 5) was found
3.14 not found!

# Pythonic way to do it:
for key in tests: # For all keys
    if key in items: # Let Python check for a match
        print(key, "was found")
    else:
        print(key, "not found!")

(4, 5) was found
3.14 not found!
```

```python
# whwnever possible use for instead of while. For is quiker
# mix for with the use of range function:
>>> list(range(5)), list(range(2, 5)), list(range(0, 10, 2)) # low, upper -1, and step
([0, 1, 2, 3, 4], [2, 3, 4], [0, 2, 4, 6, 8])

>>> list(range(−5, 5))              # allows negaive low boundaries
[−5, −4, −3, −2, −1, 0, 1, 2, 3, 4]

>>> list(range(5, −5, −1)) # allows reverse step: from positive to negative!! 
[5, 4, 3, 2, 1, 0, −1, −2, −3, −4]

# Range techinques:

>>> X = 'spam'
>>> for item in X: print(item, end=' ') # Simple iteration
...
s p a m

>>> i = 0 # called counter-loop
>>> while i < len(X): # while loop iteration --> same as above but taking over indexing logic
... print(X[i], end=' ')
... i += 1
...
s p a m

>>> X  # same example as above but using for
'spam'
>>> len(X) # Length of string
4
>>> list(range(len(X))) # All legal offsets into X
[0, 1, 2, 3]
>>>
>>> for i in range(len(X)): print(X[i], end=' ') # Manual for indexing
...
s p a m
```

_Non-exhaustive traversals_

```python
# Under those circumstances, alwas try to use the simple iterator: using for!!!!
>>> for item in X: print(item) # Simple iteration

# Specialized sorts of traversal using range:

>>> S = 'abcdefghijk'
>>> list(range(0, len(S), 2))
[0, 2, 4, 6, 8, 10]

>>> for i in range(0, len(S), 2): print(S[i], end=' ') # Only prints the step 2 in the range!!
...
a c e g i k

# the same as above but using slicing than range techinique:
>>> S = 'abcdefghijk'
>>> for c in S[::2]: print(c, end=' ') # te string (copy) is sliced with step 2
...
a c e g i k #this one is easier than range , but uses more memory: makes a copy of the str
```

_Changing lists_

```python
#incorret way to make a in-place change in a list on each element:
>>> L = [1, 2, 3, 4, 5]
>>> for x in L:   # we are changing the variable x a each iteration, but not the L[] element
... x += 1
...
>>> L
[1, 2, 3, 4, 5]
>>> x
6

# rigth way to do it using range:
>>> L = [1, 2, 3, 4, 5]
>>> for i in range(len(L)): # Add one to each item in L
... L[i] += 1 # Or L[i] = L[i] + 1
...
>>> L
[2, 3, 4, 5, 6]

# Equivalent while loop

>>> i = 0
>>> while i < len(L):
... L[i] += 1
... i += 1
...
>>> L
[3, 4, 5, 6, 7]

# Equivalent using list comprehension:

L = [1, 2, 3, 4, 5]
L = [x+1 for x in L]

```

___Parallel Traversals: zip and map___
- ZIP: Allows us to traverse many lists (foor loop) simultanously (in paralell)

```python
# ZIP:

>>> L1 = [1,2,3,4]
>>> L2 = [5,6,7,8]

>>> zip(L1, L2)
<zip object at 0x026523C8> # returns a Zip iterable object
>>> list(zip(L1, L2)) # list() required in 3.0, not 2.6 --> remeber to use list  or dicts to show iterable objects
[(1, 5), (2, 6), (3, 7), (4, 8)] # makes a list of paired tuples (paralell)

# mixing Zip with for loop: Paralell looping in both lists(or more...)
for (x, y) in zip(L1, L2): # uses tuple unpacking !!!
    print(x, y, '--', x+y)
...
1 5 -- 6  # in this first loop: (x,y) = (1,5)
2 6 -- 8
3 7 -- 10
4 8 -- 12

# Zip accepts any iterable object(including files):
>>> T1, T2, T3 = (1,2,3), (4,5,6), (7,8,9)
>>> T3
(7, 8, 9)
>>> list(zip(T1, T2, T3))
[(1, 4, 7), (2, 5, 8), (3, 6, 9)] #N array tuple x N arguments list as a result

# Zip when the iterable objects have different lengths:

>>> S1 = 'abc'
>>> S2 = 'xyz123'
>>>
>>> list(zip(S1, S2))
[('a', 'x'), ('b', 'y'), ('c', 'z')] # Zip matches the shortest sequences: in this case 0,1,2 positions !! 
```

```python
# Making dictionaries with Zip:
#1. making the dict by two options:

>>> D1 = {'spam':1, 'eggs':3, 'toast':5} # by literal
>>> D1
{'toast': 5, 'eggs': 3, 'spam': 1}

>>> D1 = {} # by assignment
>>> D1['spam'] = 1
>>> D1['eggs'] = 3
>>> D1['toast'] = 5

# using zip: old-fashiones style
>>> keys = ['spam', 'eggs', 'toast']
>>> vals = [1, 3, 5]

>>> list(zip(keys, vals))  # create the iterable object zip
[('spam', 1), ('eggs', 3), ('toast', 5)]
>>> D2 = {} # set an empty dict
>>> for (k, v) in zip(keys, vals): D2[k] = v # traverse the zip object and assign key and value for the dict
...
>>> D2
{'toast': 5, 'eggs': 3, 'spam': 1}


# Using dict: Pythonic way using built-in function dict()
>>> keys = ['spam', 'eggs', 'toast']
>>> vals = [1, 3, 5]
>>> D3 = dict(zip(keys, vals))
>>> D3
{'toast': 5, 'eggs': 3, 'spam': 1}
```

```python
# Map:takes a function and one or more sequence arguments and collects
#the results of calling the function with parallel items taken from the sequence(s).
map(fun, iter)
fun : It is a function to which map passes each element of given iterable.
iter : It is a iterable which is to be mapped.

>>> list(map(ord, 'spam')) # applies function ord to each letter in the str sequence
[115, 112, 97, 109]

# same result using for:
>>> res = []
>>> for c in 'spam': res.append(ord(c))
>>> res
[115, 112, 97, 109] # but it is slower

```

Generating Both Offsets and Items: enumerate

```python
# traditional way to carry an offset:
>>> S = 'spam'
>>> offset = 0
>>> for item in S:
... print(item, 'appears at offset', offset)
... offset += 1
...
s appears at offset 0  # its like carring every single element of rancge as the loopp continues
p appears at offset 1
a appears at offset 2
m appears at offset 3

# Using enumerate() build-in function: simplier way!
>>> S = 'spam'
>>> for (offset, item) in enumerate(S): # enumerate is a generator object: has a next method that returns the tuple (index,value)
... print(item, 'appears at offset', offset) # at each iteration through the loop
...
s appears at offset 0 # offset,item = enumerate(S[0]) = (0,s)
p appears at offset 1
a appears at offset 2
m appears at offset 3

# enumerate at length:
>>> E = enumerate(S)
>>> E
<enumerate object at 0x02765AA8> # creates a iterableobject
>>> next(E)
(0, 's')
>>> next(E)
(1, 'p')
>>> next(E)
(2, 'a')

# using list comprehension and enumerate:
>>> [c * i for (i, c) in enumerate(S)]
['', 'p', 'aa', 'mmm']
```
