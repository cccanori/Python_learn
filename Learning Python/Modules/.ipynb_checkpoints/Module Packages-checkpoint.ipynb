___Module Packages:___   An import can name with a directory path (Package) 

```python
# In the way you set the names of modules, you can set a directory path names and then the module:

import dir1.dir2.mod #mod is a subfile dof dir2, which is a subfile of dir1
#or
from dir1.dir2.mod import x
# In windows, it looks like this: dir0\dir1\dir2\mod.py # Or mod.pyc, mod.so, etc.
# where dir0 is the PYTHONPATH defined settings (sys.path)

'''common mistake:'''
import C:\mycode\dir1\dir2\mod # Error: illegal syntax
# Supposing that C:\mycode is your PYTHONPATH, then the syntaxis is as following:
import dir1.dir2.mod
```

__Package __ __init___ __.py__ Files

```python
# every single directory path must have a __init__.py attribute or the import will fail.
'''dir0\dir1\dir2\mod.py which in import terms is:'''
import dir1.dir2.mod
#the above statment has the noxt rules:
''' • dir1 and dir2 both must contain an __init__.py file.
    • dir0, the container, does not require an __init__.py file; this file will simply be
      ignored if present.
    • dir0, not dir0\dir1, must be listed on the module search path sys.path.'''
#__init__.py should be created !!! they could be empty or filled with coding. The import statment will execute the statments at first!

#Don’t confuse package __init__.py files with the class __init__ constructor methods !!!
%------------ Example------------------------------
# Coding in each directory and the main module:
# dir1\__init__.py
print('dir1 init')
x = 1

# dir1\dir2\__init__.py
print('dir2 init')
y = 2

# dir1\dir2\mod.py
print('in mod.py')
z = 3

# Then the calling:
>>> import dir1.dir2.mod # First imports run init files
dir1 init # Run in dir1's container directory
dir2 init
in mod.py
>>>
>>> import dir1.dir2.mod # Later imports do not --> it is already loaded

# Reload can be used as well
>>> from imp import reload # from needed in 3.X only
>>> reload(dir1)
dir1 init
<module 'dir1' from '.\\dir1\\__init__.py'>
>>>
>>> reload(dir1.dir2) #Reload statment accepts qualification (object.attribute)
dir2 init
<module 'dir1.dir2' from '.\\dir1\\dir2\\__init__.py'>

#Then, directories become hierarchically nested objects in the script:
>>> dir1
<module 'dir1' from '.\\dir1\\__init__.py'>
>>> dir1.dir2
<module 'dir1.dir2' from '.\\dir1\\dir2\\__init__.py'>
>>> dir1.dir2.mod
<module 'dir1.dir2.mod' from '.\\dir1\\dir2\\mod.py'>

# and eah nested object has its respective functions as qualifications:
>>> dir1.x
1
>>> dir1.dir2.y
2
>>> dir1.dir2.mod.z
3
```

__from Versus import with Packages__

```python
'''It’s often more convenient, therefore, to use the from statement with packages to avoid
retyping the paths at each access.
    
    If you ever restructure your directory tree, the from statement requires just one path update in your code,
whereas imports may require many'''

>>> from dir1.dir2 import mod # Code path here only
dir1 init
dir2 init
in mod.py
>>> mod.z # Don't repeat path --> Smaller typing !
3
>>> from dir1.dir2.mod import z #imprts directly the variable(can clash with the top-level script)
>>> z
3
>>> import dir1.dir2.mod as mod # Use shorter name (see Chapter 25) --> aliasing it with as
>>> mod.z
3
>>> from dir1.dir2.mod import z as modz # Ditto if names clash (see Chapter 25) --> again aliasing the variable so to avoid any clash
>>> modz
3
```

```python
# Why use packages ?
'''* Make imports more informative
   * Srve as a organizational tool
   * A code more readable -> shows where the modules come from
   * Help to solve ambiguities among the modules you import (modules with the same name) '''
# e.g, the first option gives more information than the second one:
import database.client.utilities
import utilities
```

__Relative Import__

```python
'''In both Python 3.X and 2.X, from statements can now use leading dots (“.”) to specify
   that they require modules located within the same package (known as package relative
   imports), instead of modules located elsewhere on the module import search path
   (called absolute imports).'''
''' In both Python 3.X and 2.X, from statements can now use leading dots (“.”) to specify
    that they require modules located within the same package (known as package relative
    imports), instead of modules located elsewhere on the module import search path
    (called absolute imports).'''

from . import spam # Relative to this package 
#instructs Python to import a module named spam located in the same package directory as the file in which this statement appears

from .spam import name # “from a module named spam located in the same package as the file that contains this statement, import the variable name.”

#Comparison:
import string # Skip this package's version --> looks for the module skip somewhere on the sys.path
from . import string # Searches this module in the current package only

#Extenions:assuming mypkg is the package directory
from .string import name1, name2 # Imports names from mypkg.string
from . import string # Imports mypkg.string
from .. import string # Imports string sibling of mypkg

'''ode located in some module A.B.C can use any of these forms:''' # Actually the working script is in C
from . import D # Imports A.B.D (. means A.B) -> 
from .. import E # Imports A.E (.. means A)
from .D import X # Imports A.B.D.X (. means A.B)
from ..E import X # Imports A.E.X (.. means A)
%---------------------------------------------------------------------
# Example:
'''code\string.py'''
print('string' * 8)

'''code\pkg\spam.py''' # Assume that __init__.py is already in the pkg
from . import eggs
print(eggs.X)

'''code\pkg\eggs.py''' # Assume that __init__.py is already in the pkg
X = 99999
import string # <== Gets string in CWD, not Python lib!
print(string)

C:\code> c:\Python33\python # Same result in 2.X
>>> import pkg.spam
stringstringstringstringstringstringstringstring
<module 'string' from '.\\string.py'>
99999
%------------------------------------------------------------------------------------
# be aware of:
from . import mod # Not allowed in nonpackage mode in both 2.X and 3.X
import mod # Does not search file's own directory in package mode in 3.X
```

__Python 3.3 Namespace Packages__

```python
'''*Basic module imports: import mod, from mod import attr
        The original model: imports of files and their contents, relative to the sys.path
        module search path
    *Package imports: import dir1.dir2.mod, from dir1.mod import attr
        Imports that give directory path extensions relative to the sys.path module search
        path, where each package is contained in a single directory and has an initialization
        file, in Python 2.X and 3.X
    *Package-relative imports: from . import mod (relative), import mod (absolute)
        The model used for intrapackage imports of the prior section, with its relative or
        absolute lookup schemes for dotted and nondotted imports, available but differing
        in Python 2.X and 3.X
    *Namespace packages: import splitdir.mod
        The new namespace package model that we’ll survey here, which allows packages
        to span multiple directories, and requires no initialization file, introduced in Python
        3.3'''
'''In terms of physical structure, though, the two can differ substantially. Regular packages
    still must have an __init__.py file that is run automatically, and reside in a single
    directory as before. By contrast, new-style namespace packages cannot contain an
    __init__.py, and may span multiple directories that are collected at import time.'''

```
