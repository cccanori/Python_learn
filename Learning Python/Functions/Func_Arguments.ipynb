___ARGUMENTS PASSING___ (The way objects are send to functions as inputs)

```python
# Arguments are passed by assignments --> means that functions only accept pointers not objects directly!!
# Arguments passed in the header of the function become new local names: local scope variables
# Functions can change passed mutable objects arguments: the result may affect the caller
# passing arguments with inmutable objects is fair similar t make a copy of the object
```

__Arguments and Shared References__

```python
# ex with inmutable object number: doesn't change the object, and then, does not change the pointer

def f(a): # a is assigned to (references) the passed object
    a = 99 # Changes local variable a only

b = 88
f(b) # a and b both reference same 88 initially -> then -> a refrences 99 but in the local scope without chanching b 
print(b) # b is not changed -> variable a only lives wihin the def function -> does not change b
88

#Above example at length by assignments:
>>> X = 1
>>> a = X # They share the same object
>>> a = 2 # Resets 'a' only, 'X' is still 1
>>> print(X)
1
#--------------------------------------------------------------------------

#example with inmutable (number) and mutable (list) objects: the latter is changed because the object is changed!
def changer(a, b): # Arguments assigned references to objects
    a = 2 # Changes local name's value only
    b[0] = 'spam' # Changes shared object in place --> doesn't change b, only part of the object it  references
X = 1
L = [1, 2] # Caller:
changer(X, L) # Pass immutable and mutable objects
X, L # X is unchanged, L is different!
(1, ['spam', 2])

# above example at length by assignments:
>>> L = [1, 2]
>>> b = L # They share the same object
>>> b[0] = 'spam' # In-place change: 'L' sees the change too
>>> print(L)
['spam', 2]
#------------------------------------------------------------------------------
# solution for avoid changing mutable obejcts: use copies
L = [1, 2]
changer(X, L[:]) # Pass a copy, so our 'L' does not change
#or
def changer(a, b):
    b = b[:] # Copy input list so we don't impact caller
    a = 2
    b[0] = 'spam' # Changes our list copy only

# or the more strict (then less flexible) tuple: avoid this one if it is possible!
L = [1, 2]
changer(X, tuple(L)) # Pass a tuple, so changes are errors --> make sure that your iterable object isn't changed
```

__Simulating Output Parameters and Multiple Results__

```python
# return can retrive multiple objects by using a packing collection type:
def multiple(x, y):
    x = 2 # Changes local names only
    y = [3, 4]
    return x, y # Return multiple new values in a tuple

>>> X = 1
>>> L = [1, 2]
>>> X, L = multiple(X, L) # Assign results to caller's names --> global assignment by unpacking the tuple!!
>>> X, L
(2, [3, 4])
```

__Special Argument-Matching Modes__

```python
# by default arguments match objects bay position (left to rigth) --> However, there are some exceptions: name, provide default, and collections
# Arguments basics:
- Positionals: matched from left to right: the normal one
- Keywords: matched by argument name --> name=value syntax.
- Defaults: specify values for optional arguments that arenâ€™t passed --> again using the name=value syntax.
- Varargs collecting: collect arbitrarily many positional or keyword arguments --> special arguments preceded with * or ** characters to collect an arbitrary number of possibly extra arguments.
- Varargs unpacking: pass arbitrarily many positional or keyword arguments --> use the * syntax to unpack argument collections into separate arguments. This is the inverse of a * in a function header
- Keyword-only arguments: arguments that must be passed by name --> functions can also specify arguments that must be passed by name with keyword arguments, not by position.

# Basically: default is used when you pass less arguments than the ones required and * allows you to pass more arguments.
# when * is used, the excess of variables is collected acording to the function
```

```python
#Ordering rules- callas:
1. positional arguments(value) 2.keywords arguments (name= value) 3. *iterable form 4.**dict form
# Ordering rules - def:
1. name 2. name = value (default) 3.*name 4. name = value (keyword-only-argument) 5.**name

#match by positional arguments:
def f(a, b, c): print(a, b, c)
f(1, 2, 3) # python matches 123 with abc respectively
1 2 3
%------------------------------------------------------
# match by names:
f(c=3, b=2, a=1)  # left to rigth order doesnt matter any more
1 2 3
%-------------------------------------------------------------
# match by position and name:
f(1, c=3, b=2) # a gets 1 by position, b and c passed by name --> also, position arguments are first, then the name arguments!
1 2 3

# using name= values is widely more informative than values divided by commas:
func(name='Bob', age=40, job='dev') has more meaning than func('Bob',40,'dev').
%-------------------------------------------------------------------------------

#Defaults:

>>> def f(a, b=2, c=3): print(a, b, c) # a required, b and c optional
>>> f(1) # Use defaults
1 2 3
>>> f(a=1)
1 2 3
>>> f(1, 4) # Override defaults --> c is default (left to rigth)
1 4 3
>>> f(1, 4, 5) # no defaults at all
1 4 5
>>> f(1, c=6) # Choose defaults
1 2 6
# be caful: neither the header ot the call is soing an assignment, they are just special cases of syntaz regarding argument-matching

#Using defaults and matching (position and names):
def func(spam, eggs, toast=0, ham=0): # First 2 required
    print((spam, eggs, toast, ham))

func(1, 2) # Output: (1, 2, 0, 0) #---> spam and egs are matched but toast and ham are default values
func(1, ham=1, eggs=0) # Output: (1, 0, 0, 1)
func(spam=1, eggs=0) # Output: (1, 0, 0, 0)
func(toast=1, eggs=2, spam=3) # Output: (3, 2, 1, 0)
func(1, 2, 3, 4) # Output: (1, 2, 3, 4)
# the assignment '=' means a different think in the caller (keyword) and the def header (default)
'''be carfull of mutable defaults (e.g., f(a=[])) !! they are re used at each function call--> not reset to the original
    then, very sensible to in-place changes--> retain the status of previous calls --> Hence, use the mutable object in
    the body of the function instead of the header --> another type of state retention'''
%-----------------------------------------------------------------------------------------------
'UNPACKETING' -> CALL (UNPACKING TUPLES * AND DICTS**) 'PACKETING' -> DEF(PACKING TUPLES * AND DICTS**)
# arbitrary arguments: * and ** support function with any amount of arguments in both the caller and the header(def)
HEADERS*:
def f(*args): print(args) # collect the positional arguments into a tuple!!! --> value assignment
>>> f()
()    # empty tuple
>>> f(1) # tuple with one element 
(1,)
>>> f(1, 2, 3, 4)
(1, 2, 3, 4) #tuple with 4 eleemnts

HEADERS**:
def f(**args): print(args) # collect the arguments into a dict --> must y name=value form!!!
f()
{}
f(a=1, b=2) # name=value (keyword) asssignment!!!
{'a': 1, 'b': 2}

HEADERS * and **:
>>> def f(a, *pargs, **kargs): print(a, pargs, kargs)
>>> f(1, 2, 3, x=1, y=2) # 1 is assigned by left-to-rigth assignemnt, 2 and 3 are assigned as tuple(value) and x and y wind up in a dict (name=value) 
1 (2, 3) {'y': 2, 'x': 1}

>>> def echo(*args, **kwargs): print(args, kwargs)
>>> echo(1, 2, a=3, b=4)
(1, 2) {'a': 3, 'b': 4}


CALLS: *
>>> def func(a, b, c, d): print(a, b, c, d)
>>> args = (1, 2)
>>> args += (3, 4)
>>> func(*args) # Same as func(1, 2, 3, 4) --> unpackes the tuple as in variable assignment
1 2 3 4

CALLS: **
>>> args = {'a': 1, 'b': 2, 'c': 3}
>>> args['d'] = 4
>>> func(**args) # Same as func(a=1, b=2, c=3, d=4) --> unpackes a dict in a name=value tupple
1 2 3 4

CALLS: * and **
>>> func(*(1, 2), **{'d': 4, 'c': 3}) # Same as func(1, 2, d=4, c=3)
1 2 3 4
>>> func(1, *(2, 3), **{'d': 4}) # Same as func(1, 2, 3, d=4)
1 2 3 4
>>> func(1, c=3, *(2,), **{'d': 4}) # Same as func(1, 2, c=3, d=4)
1 2 3 4
>>> func(1, *(2, 3), d=4) # Same as func(1, 2, 3, d=4)
1 2 3 4
>>> func(1, *(2,), c=3, **{'d':4}) # Same as func(1, 2, c=3, d=4)
1 2 3 4

>>> def echo(*args, **kwargs): print(args, kwargs) #this function packes the values in a tuple and the keywords(name=value) in a dict 
>>> pargs = (1, 2)
>>> kargs = {'a':3, 'b':4}
>>> echo(0, c=5, *pargs, **kargs) # Normal, keyword, *sequence(unpacking), **dictionary(unpacking)
(0, 1, 2) {'a': 3, 'c': 5, 'b': 4} #the function returns the arguments packed again

#REMEMBER: */** ARE POSITIONAL AND KEY WORDS--> THE FORMER PACKS(HEADER) AND THE LATTER UNPACKS(CALLER)

%-----------------------------------------------------------------------------
#applying function generically:
# Quite useful when you dont know the function or arguemnts that are finally required
if sometest:
action, args = func1, (1,) # Call func1 with one arg in this case --> action = func1 and arg =(1,)
else:
action, args = func2, (1, 2, 3) # Call func2 with three args here
...etc...
action(*args) # Dispatch generically--> eg case 1: func1(*(1,)) = action(*args)
# assigns the function and the arguments depending the if clause result

#the following code we support any function with any arguments by passing along whatever arguments were sent in:
def tracer(func, *pargs, **kargs): # Accept arbitrary arguments
    print('calling:', func.__name__)
    return func(*pargs, **kargs) # Pass along arbitrary arguments
def func(a, b, c, d):
    return a + b + c + d

print(tracer(func, 1, 2, c=3, d=4))
calling: func
10


```

__Example function in practice:__ Coding a function that take a set of objectes (any) and returns the minimum value

```python
#The three examples uses polymorphism --> let python take care of adapt the oprand given a certain object
def min1(*args): # packs the arguments(all) into a tuple 
    res = args[0] #saves the first one in a list
    for arg in args[1:]: #traverses the list and compare the others[1:] against the first one[0]
        if arg < res:
            res = arg  #change the pointer if True
    return res

def min2(first, *rest): #save the first one by index and again traverse the rest and change the pointer in case of finding a True
    for arg in rest:
        if arg < first:
        first = arg
    return first

def min3(*args): #Pack al arguments in a tuple
    tmp = list(args) # Or, in Python 2.4+: return sorted(args)[0] --> transform them in a list
    tmp.sort()  #let sort method and polymorphism work
    return tmp[0] #retrun the first object

print(min1(3, 4, 1, 2))
print(min2("bb", "aa"))
print(min3([2,2], [1,1], [3,3]))

1
aa
[1, 1]

# Coding a function that evaluates the min or Max depending on the situation:

def minmax(test, *args): #test assign references a forward function
    res = args[0]
    for arg in args[1:]:
        if test(arg, res):# call the function-> nested function--> returns True or False
        res = arg
    return res
def lessthan(x, y): return x < y # See also: lambda, eval
def grtrthan(x, y): return x > y

print(minmax(lessthan, 4, 2, 1, 5, 6, 3)) # Self-test code
print(minmax(grtrthan, 4, 2, 1, 5, 6, 3))
1
6
# this exercise do the same as min() and max() nuilt in functions

'''Tester function: uses function* assignemnt to have a traceback'''
# the two functions we want to test : a intersection function and union function (among any iterable objects)

def intersect(*args): #packs all positional arguments in a tuple
    res = []
    for x in args[0]: # Scan first sequence
        if x in res: continue # Skip duplicates
        for other in args[1:]: # For all other args
            if x not in other: break # Item in each one?
        else: # No: break out of loop
            res.append(x) # Yes: add items to end
    return res

def union(*args):
    res = []
    for seq in args: # For all args
        for x in seq: # For all nodes
            if not x in res:
                res.append(x) # Add new items to result
    return res
# functions in action:
>>> from inter2 import intersect, union
>>> s1, s2, s3 = "SPAM", "SCAM", "SLAM"
>>> intersect(s1, s2), union(s1, s2) # Two operands
(['S', 'A', 'M'], ['S', 'P', 'A', 'M', 'C'])
>>> intersect([1, 2, 3], (1, 4)) # Mixed types
[1]
>>> intersect(s1, s2, s3) # Three operands
['S', 'A', 'M']
>>> union(s1, s2, s3)
['S', 'P', 'A', 'M', 'C', 'L']

# Setting the tester function:
def tester(func, items, trace=True):
    for i in range(len(items)): #traverse all itmes
        items = items[1:] + items[:1] #shuffle it sending the first one to the las position --> uses a copy and assigns the pointer
        if trace: print(items) #by default prints each shuffle
        print(sorted(func(*items))) #uses sorted method(polymorphism) in the output of the evaluated function to asses the final result

#testing it
>>> tester(intersect, ('a', 'abcdefg', 'abdst', 'albmcnd'))
('abcdefg', 'abdst', 'albmcnd', 'a') # a is shuffled from the first position to the last until it reaches again the first position
['a']
('abdst', 'albmcnd', 'a', 'abcdefg')
['a']
('albmcnd', 'a', 'abcdefg', 'abdst')
['a']
('a', 'abcdefg', 'abdst', 'albmcnd')
['a']
>>> tester(union, ('a', 'abcdefg', 'abdst', 'albmcnd'), False)
['a', 'b', 'c', 'd', 'e', 'f', 'g', 'l', 'm', 'n', 's', 't']
['a', 'b', 'c', 'd', 'e', 'f', 'g', 'l', 'm', 'n', 's', 't']
['a', 'b', 'c', 'd', 'e', 'f', 'g', 'l', 'm', 'n', 's', 't']
['a', 'b', 'c', 'd', 'e', 'f', 'g', 'l', 'm', 'n', 's', 't']
>>> tester(intersect, ('ba', 'abcdefg', 'abdst', 'albmcnd'), False)
['a', 'b']
['a', 'b']
['a', 'b']
['a', 'b']

>>> intersect([1, 2, 1, 3], (1, 1, 4))
[1]
>>> union([1, 2, 1, 3], (1, 1, 4))
[1, 2, 3, 4]
>>> tester(intersect, ('ababa', 'abcdefga', 'aaaab'), False)
['a', 'b']
['a', 'b']
['a', 'b']

# the tester could be even simplified usinga function that scramble the items

def tester(func, items, trace=True):
    for args in scramble(items):
        ...use args...
# this function is kind of obsolete now that we can use the object set to perfomr those operations


'''Emulating python 3.X print'''
#!python print3 --> name of the saved file
"""
Emulate most of the 3.X print function for use in 2.X (and 3.X).
Call signature: print3(*args, sep=' ', end='\n', file=sys.stdout)
"""
import sys
def print3(*args, **kargs):
    sep = kargs.get('sep', ' ') # Keyword arg defaults
    end = kargs.get('end', '\n')
    file = kargs.get('file', sys.stdout)
    if kargs: raise TypeError('extra keywords: %s' % kargs) # if more name=value assigntments are passed than the expected we raise an error --> control code
    output = ''
    first = True
    for arg in args:
        output += ('' if first else sep) + str(arg)
        first = False
    file.write(output + end)

#Function in action:
from print3 import print3 # calling the made function by loading the module
print3(1, 2, 3)
1 2 3
print3(1, 2, 3, sep='') # Suppress separator
123
print3(1, 2, 3, sep='...')
1...2...3
print3(1, [2], (3,), sep='...') # Various object types
1...[2]...(3,)
print3(4, 5, 6, sep='', end='') # Suppress newline
print3(7, 8, 9)
print3() # Add newline (or blank line)
4567 8 9

import sys
print3(1, 2, 3, sep='??', end='.\n', file=sys.stderr) # Redirect to file
1??2??3.
```
