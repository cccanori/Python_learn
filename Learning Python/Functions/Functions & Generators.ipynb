__Polymorphism in Python__

```python
# a function (x*y) can perform multiplication in one instance and repetition in another.
# All operations in python are polymorphism --> depend on the object arguments
# python will returns an error if the objects(mix of different) do not support the operand (e.g, *)

#¡¡ In python you are not supposed to care about data types!! --> we code object interfaces, no data types
# reduces amount of code and increases flexibility

def intersect(seq1, seq2):
    res = [] # Start empty
    for x in seq1: # Scan seq1
        if x in seq2: # Common item?
            res.append(x) # Add to end
return res

>>> s1 = "SPAM"
>>> s2 = "SCAM"
>>> intersect(s1, s2) # Strings
['S', 'A', 'M']

>>> x = intersect([1, 2, 3], (1, 4)) # Mixed types --> Polymorphism!! tuple and list support the operand in the function!
>>> x # Saved result object
[1]


```

__SCOPES:__ Names defined in one program unit don’t interfere with names in another. And retain state information within calls

```python
# namespace determines the name assignment in your code --> scope of visibility
# all variables assigned within a function are associated with the function namespace--> are only valid within the function!

#Lexical scoping:
• If a variable is assigned inside a def, it is local to that function.
• If a variable is assigned in an enclosing def, it is nonlocal to nested functions.
• If a variable is assigned outside all defs, it is global to the entire file.

# variables' scope is determined by the location of them in the code, not by the call expression!!!

X = 99 # Global (module) scope X
def func():
    X = 88 # Local (function) scope X: a different variable

```

- The enclosing module is a global scope.
- The global scope spans a single file only.
- Assigned names are local unless declared global or nonlocal.
- All other names are enclosing function locals, globals, or built-ins.
- Each call to a function creates a new local scope. --> Recursion: funct calling itself -->own copy of funct local-var.
- In-plache chamges (Methods) affect the object(global), then affect the variables even within a def function!

```python
# The LEGB rule:
When you use an unqualified name inside a function, Python searches up to four
scopes—the local (L) scope, then the local scopes of any enclosing (E) defs and
lambdas, then the global (G) scope, and then the built-in (B) scope—and stops at
the first place the name is found. If the name is not found during this search, Python
reports an error.
```

__Scopes and Nested Functions__
- __A reference (X)__ looks for the name X first in the current local scope (function); then
in the local scopes of any lexically enclosing functions in your source code, from
inner to outer; then in the current global scope (the module file); and finally in the
built-in scope (the module builtins). global declarations make the search begin
in the global (module file) scope instead.
- __An assignment (X = value)__ creates or changes the name X in the current local
scope, by default. If X is declared global within the function, the assignment creates
or changes the name X in the enclosing module’s scope instead. If, on the other
hand, X is declared nonlocal within the function in 3.X (only), the assignment
changes the name X in the closest enclosing function’s local scope.

```python
# Nested scope example
X = 99 # Global scope name: not used
def f1():
    X = 88 # Enclosing def local
    def f2():
        print(X) # Reference made in nested def --> Local
    f2()
f1() # prints 88 because it makes reference to th X local variable in f1 -> enclosed in the def nest
```

```python
def f1():
    X = 88
    def f2():
        print(X) # Remembers X in enclosing def scope
    return f2 # Return f2 but don't call it -> return of f1 not f2!

action = f1() # Make, return function --> saving the return in action --> action = f2
action() # Call it now: prints 88 -->  execute action(fi()) without saving the return
```

__Factory Functions: Closures__ (functional programming and design pattern respectively))

```python
# creating a aker and action function: nested functions!!
def maker(N): # defining the factory function
    def action(X): # Make and return action
        return X ** N # action retains N from enclosing scope
    return action

f = maker(2) # Pass 2 to argument N --> setting the maker function
f
<function maker.<locals>.action at 0x0000000002A4A158>

>>> f(3) # Pass 3 to X, N remembers 2: 3 ** 2 --> evaluating 3 at the action function while the maker function argument is 2
9
>>> f(4) # 4 ** 2
16

# Generating a new maker (now we have f and g):
g = maker(3) # g remembers 3, f remembers 2
g(4) # 4 ** 3 -> using maker(3)
64
f(4) # 4 ** 2 --> using maker (2)
16

#In our case, the function we assign to name g remembers 3, and f remembers 2, 
#because each has its own state information retained by the variable N in maker.

# Same as above using lambda: Lamda is a creting-function expression, could be nested in the def as well

def maker(N):
    return lambda X: X ** N # lambda functions retain state too
h = maker(3) #saving the paremeter 3 in the factory function
h(4) # 4 ** 3 again --> #evaluating the action function (lamda X) in 4
64
# Quite popular in functional programming --> the goal is to retaining state !
# the factory function provide a type of 'class' for the nested function 
```

__Loop variables may require defaults, not scopes__

```python
# Generating a function that buils up a listfrom a loop at eahciteration of i:
def makeActions():
    acts = []
    for i in range(5): # Tries to remember each i as elements of list 'acts'
        acts.append(lambda x: i ** x) # But all remember same last i!
    return acts
acts = makeActions() #save the makeactions() factory function..
acts[0] # retrive the first element of actions lists generated by the function
<function makeActions.<locals>.<lambda> at 0x0000000002A4A400>

>>> acts[0](2) # All are 4 ** 2, 4=value of last i --> retrives the firs element of list acts after running the paraemte 2 in the factory function
16
>>> acts[1](2) # This should be 1 ** 2 (1)
16
>>> acts[2](2) # This should be 2 ** 2 (4)
16
>>> acts[4](2) # Only this should be 4 ** 2 (16)
16
# all are 16 evnethough they should be different -> the function only remebers the las iteration i !

# Solution: explicitely retain enclosing scopes with default arguments:
def makeActions():
    acts = []
    for i in range(5): # Use defaults instead
        acts.append(lambda x, i=i: i ** x) # Remember current i
    return acts
acts = makeActions()

>>> acts[0](2) # 0 ** 2
0
>>> acts[1](2) # 1 ** 2
1
>>> acts[2](2) # 2 ** 2
4
>>> acts[4](2) # 4 ** 2
16
```

__The nonlocal Statement in 3.X__ (nonlocal declares that a name will be changed in an enclosing scope)
- Unlike global, though, nonlocal applies to a name in an enclosing function’s scope, not the global module scope outside all defs.
- enclosed variables in def function can look up variables in the nested def functions (E) but cannot change them, with the non-local they can ! -> writable state of information
- non-local also means “skip my local scope entirely.”

```python
# general form:
def func():
    nonlocal name1, name2, ... # OK here
>>> nonlocal X
SyntaxError: nonlocal declaration not allowed at module level

• global makes scope lookup begin in the enclosing module’s scope and allows
    names there to be assigned. Scope lookup continues on to the built-in scope if the
    name does not exist in the module, but assignments to global names always create
    or change them in the module’s scope.

• nonlocal restricts scope lookup to just enclosing defs, requires that the names already
    exist there, and allows them to be assigned. Scope lookup does not continue
    on to the global or built-in scopes.
```

```python
# Normal scope rule:
def tester(start):
    state = start # Referencing nonlocals works normally
    def nested(label):
        print(label, state) # Remembers state in enclosing scope
    return nested

F = tester(0) # set the maker function (start argument) at 0 in the pointer F
F('spam') #evaluate the nested function with the argument label 
spam 0

F('ham')
ham 0

# trying to change state in the nested def scope:
def tester(start):
    state = start
    def nested(label):
        print(label, state)
        state += 1 # Cannot change by default (never in 2.X) --> state could be referenced but not changed !!
    return nested
>>> F = tester(0)
>>> F('spam')
UnboundLocalError: local variable 'state' referenced before assignment

# Solution: Using nonlocal for changes !
def tester(start):
    state = start # Each call gets its own state
    def nested(label):
        nonlocal state # Remembers state in enclosing scope --> Now state could be modified !!
        print(label, state)
        state += 1 # Allowed to change it if nonlocal
    return nested

>>> F = tester(0)
>>> F('spam') # Increments state on each call
spam 0
>>> F('ham')
ham 1
>>> F('eggs')
eggs 2

# Changing the start of the list does not affect the pattern of the previous: G won't affect F
>>> G = tester(42) # Make a new tester that starts at 42
>>> G('spam')
spam 42
>>> G('eggs') # My state information updated to 43
eggs 43
>>> F('bacon') # But F's is where it left off: at 3
bacon 3 # Each call has different state information --> because they are not global but non-local -enclosed in def not in the module
```

__State with nonlocal: 3.X only__

```python

# Each call to tester() creates a self-contained (local) package of info that do not clashes with global variables:
def tester(start):
    state = start # Each call gets its own state
    def nested(label):
        nonlocal state # Remembers state in enclosing scope
        print(label, state)
        state += 1 # Allowed to change it if nonlocal
    return nested

>>> F = tester(0)
>>> F('spam') # State visible within closure only
spam 0
>>> F.state
AttributeError: 'function' object has no attribute 'state'
```

__State with Globals: A Single Copy Only__

```python
#One common prescription for achieving the nonlocal effect in 2.X and earlier is to simply move the state out to the global scope (the enclosing module):
def tester(start):
    global state # Move it out to the module to change it
    state = start # global allows changes in module scope
    def nested(label):
        global state
        print(label, state)
        state += 1
    return nested

>>> F = tester(0)
>>> F('spam') # Each call increments shared global state
spam 0
>>> F('eggs')
eggs 1

#problem is that it only allows for a single shared copy of the state information in the module scope:

>>> G = tester(42) # Resets state's single copy in global scope
>>> G('toast')
toast 42
>>> G('bacon')
bacon 43
>>> F('ham') # But my counter has been overwritten!
ham 44

#when you are using nonlocal and nested function closures instead of global, each call to tester remembers its own unique copy of the state object.
```

__State with Classes: Explicit Attributes (Preview)__

```python
#The other prescription for changeable state information in 2.X and earlier is to use
#classes with attributes to make state information access more explicit than the implicit magic of scope lookup rules.

class tester: # Class-based alternative (see Part VI)
    def __init__(self, start): # On object construction,
        self.state = start # save state explicitly in new object
    def nested(self, label):
        print(label, self.state) # Reference state explicitly
        self.state += 1 # Changes are always allowed --> self argument automatically receives the implied subject of the call
                                                           # (an instance object created by calling the class itself)
                                                           # avoids the function scope rule and look for it directly on the class

>>> F = tester(0) # Create instance, invoke __init__ --> start is 0
>>> F.nested('spam') # F is passed to self
spam 0
>>> F.nested('ham')
ham 1
# supports multiple copies of retained data: each variable with tester(x) point to different objects (classes)

>>> G = tester(42) # Each instance gets new copy of state
>>> G.nested('toast') # Changing one does not impact others
toast 42
>>> G.nested('bacon')
bacon 43
>>> F.nested('eggs') # F's state is where it left off
eggs 2
>>> F.state # State may be accessed outside class
3

# same as above Using operators overloading:
class tester:
    def __init__(self, start):
    self.state = start
    def __call__(self, label): # Intercept direct instance calls
        print(label, self.state) # So .nested() not required
        self.state += 1
>>> H = tester(99) # set start argument
>>> H('juice') # Invokes __call__
juice 99
>>> H('pancakes')
pancakes 100

# classes do not need to state their variables as 'non-local' in inner functions(exlicit attributes) 
```

__State with Function Attributes: 3.X and 2.X__
- we can also sometimes achieve the same effect as nonlocals with function attributes—user-defined names attached to functions directly.

```python
# Same code but using attributes:
def tester(start):
    def nested(label):
        print(label, nested.state) # nested is in enclosing scope
        nested.state += 1 # Change attr, not nested itself --> no need for non-local statment - like an in-place change
    nested.state = start # Initial state after func defined -> doesn't change after the function creation at least that we explicitely do that
    return nested        # is the initial attribute and remainds like that !!

>>> F = tester(0) # nested.state is 0 as we defined start as 0 !! --> F.state = x is the way to change it after is created
>>> F('spam') # F is a 'nested' with state attached
spam 0
>>> F('ham')
ham 1
>>> F.state # Can access state outside functions too --> and change it or display !
2

#Because each call to the outer function produces a new nested function object:

G = tester(42) # G has own state, doesn't overwrite F's
G('eggs')
eggs 42
F('ham')
ham 2
>>> F.state # State is accessible and per-call
3
>>> G.state
43
>>> F is G # Different function objects
False

# functions attributes make the use of non-local redundant and far less portable
```
